```python
#Step 3:Extracting the rows and columns we need

years_df =pd.DataFrame()
years_df = climate_df[["Country code","Series name",2000,2001,2002,2003,2004,2005,2006,2007,2008,2009,2010]]

years_df = years_df.loc[years_df["Country code"].isin(["AUS","CAN","CHN","CHL","IND"])]
years_df= years_df.reset_index(drop=True)


years_df.head(200)
```

```python
#step 3 continued..
series_code=["Population",
"CO2 emissions per units of GDP (kg/$1,000 of 2005 PPP $)",
"CO2 emissions, total (KtCO2)",
"Energy use per units of GDP (kg oil eq./$1,000 of 2005 PPP $)",
"GDP ($)",
"Cereal yield (kg per hectare)",
]

print(series_code)

years_df = years_df.loc[years_df["Series name"].isin(series_code)]
years_df= years_df.reset_index(drop=True)
years_df.head(200)
years_df["Series name"].count()
```

```python
#Step 4 - Groupby Series name

sn_grp1 = years_df.groupby(["Series name"])
#n_grp1 = sn_grp1.reset_index()

sn_grp1.head()


```

```python
#Step 4.1 Making individual groups
#Urban population
gp_population = sn_grp1.get_group("Population")
gp_co2_emssion_gdp= sn_grp1.get_group("CO2 emissions per units of GDP (kg/$1,000 of 2005 PPP $)")
gp_ktco2= sn_grp1.get_group("CO2 emissions, total (KtCO2)")
gp_enerygy_GDP= sn_grp1.get_group("Energy use per units of GDP (kg oil eq./$1,000 of 2005 PPP $)")
gp_GDP= sn_grp1.get_group("GDP ($)")
gp_cereal_kg= sn_grp1.get_group("Cereal yield (kg per hectare)")

```

```python
#Step 5 - working on Population group


gp_population.reset_index(drop=True)

gp_population=gp_population.rename(columns={"Series name":"SN"})

#Drop the columns Series name and the col whcih has no data

gp_population=gp_population.drop(["SN",2009,2010],axis=1)

#reset index after drop
gp_population=gp_population.set_index("Country code")

#Transpose the table
gp_population_t=gp_population.T
print(gp_population_t.head())
#Plot

#Keep the colors for all country the same
country_list=[('AUS','o','Cyan'),('CAN','^','Magenta'),('CHL','s','Blue'),('CHN','d','REd'),('IND','p','Green')]
for country,marker,color in country_list:
    plt.plot(gp_population_t[country],label=country,color=color)
    
    
#plt.plot(gp_co2_capita_t)
plt.grid()
plt.legend(loc='best')
plt.title("Population")
plt.savefig("../output/population.png")
plt.show()
```

```python
##Step 5 - working on  "Cereal yield (kg per hectare)"
print(gp_cereal_kg.head())
gp_cereal_kg.reset_index(drop=True)

gp_cereal_kg=gp_cereal_kg.rename(columns={"Series name":"SN"})

#Drop the columns Series name and the col whcih has no data

gp_cereal_kg=gp_cereal_kg.drop(["SN",2010],axis=1)

#reset index after drop
gp_cereal_kg=gp_cereal_kg.set_index("Country code")

#Transpose the table
gp_cereal_kg_t=gp_cereal_kg.T

#Plot
country_list=[('AUS','o','Cyan'),('CAN','^','Magenta'),('CHL','s','Blue'),('CHN','d','REd'),('IND','p','Green')]
for country,marker,color in country_list:
    plt.plot(gp_cereal_kg_t[country],label=country,color=color)
    
#plt.plot(gp_cereal_kg_t)
plt.legend(loc='upper left')
plt.grid()
plt.title("Cereal yield (kg per hectare)")
plt.savefig("../output/cereal_yield.png")
plt.show()
```

```python
#Step 5 - working on Energy use per units of GDP (kg oil eq./$1,000 of 2005 PPP $)

gp_enerygy_GDP.reset_index(drop=True)

gp_enerygy_GDP=gp_enerygy_GDP.rename(columns={"Series name":"SN"})

#Drop the columns Series name and the col whcih has no data

gp_enerygy_GDP=gp_enerygy_GDP.drop(["SN",2010],axis=1)

#reset index after drop
gp_enerygy_GDP=gp_enerygy_GDP.set_index("Country code")

#Transpose the table
gp_enerygy_GDP_t=gp_cereal_kg.T
print(gp_enerygy_GDP_t.head())
#Plot
country_list=[('AUS','o','Cyan'),('CAN','^','Magenta'),('CHL','s','Blue'),('CHN','d','REd'),('IND','p','Green')]
for country,marker,color in country_list:
    plt.plot(gp_enerygy_GDP_t[country],label=country,color=color)
    
#plt.plot(gp_cereal_kg_t)
plt.grid()
plt.legend(loc='upper left')
plt.title("Energy use per units of GDP (kg oil eq./$1,000 of 2005 PPP $)")
plt.savefig("../output/energyusegdp.png")
plt.show()
```

```python
#Step 5 - working on "CO2 emissions, total (KtCO2)",


gp_ktco2.reset_index(drop=True)

gp_ktco2=gp_ktco2.rename(columns={"Series name":"SN"})

#Drop the columns Series name and the col whcih has no data

gp_ktco2=gp_ktco2.drop(["SN",2010],axis=1)

#reset index after drop
gp_ktco2=gp_ktco2.set_index("Country code")

#Transpose the table
gp_ktco2_t=gp_cereal_kg.T
print(gp_ktco2_t.head())

#Plot
country_list=[('AUS','o','Cyan'),('CAN','^','Magenta'),('CHL','s','Blue'),('CHN','d','REd'),('IND','p','Green')]
for country,marker,color in country_list:
    plt.plot(gp_ktco2_t[country],label=country,color=color)
    
#plt.plot(gp_cereal_kg_t)
plt.grid()
plt.legend(loc='upper left')
plt.title("CO2 emissions, total (KtCO2)")
plt.savefig("../output/cosemission.png")
plt.show()
```
